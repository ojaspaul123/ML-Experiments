```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse_output=False)
```

```python
df_trans = ohe.fit_transform(df[['FirstName','Gender','City']])
```

```python
cols = ohe.get_feature_names_out()
```
