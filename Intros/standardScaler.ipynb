```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#
```

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('monthly_car_sales_sample.csv')
df.head()
```

```python
df.shape
```

```python
df.drop('Color',axis=1,inplace=True)
df.drop('Date',axis=1,inplace=True)
df
```

```python
df['Fuel_Type'] = df['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
df
```

```python
%%markdown
---
Train_Test_Split
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Customer_City',axis=1)
y = df['Customer_City']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size =0.33,random_state=42)
```

```python
X_train
```

```python
%%markdown
Applying StandarScaler
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Select only numeric columns
X_train_num = X_train.select_dtypes(include=['int64', 'float64'])
X_test_num = X_test.select_dtypes(include=['int64', 'float64'])

# Scale numeric data
X_train_scaled = scaler.fit_transform(X_train_num)
X_test_scaled = scaler.transform(X_test_num)
```

```python
X_train_scaled
```

```python
X_train_scal = pd.DataFrame(X_train_scaled, columns=X_train_num.columns)
X_test_scal = pd.DataFrame(X_test_scaled, columns=X_test_num.columns)
```

```python
np.round(X_train.describe(), 2)
```

```python
np.round(X_train_scal.describe(), 2)
```

```python
fig,(x1,x2) = plt.subplots(1,2,figsize=(10,5))
x1.scatter(X_train['Unit_Price'],X_train['Total_Amount'])
plt.title("Before Scalling")
x2.scatter(X_train_scal['Unit_Price'],X_train_scal['Total_Amount'],color ='red')
plt.title("After Scalling")
plt.xlabel("Unit_Price")
plt.ylabel("Total_Amount")
plt.show()
```
