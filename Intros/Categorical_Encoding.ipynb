```python
import pandas as pd
import numpy as np
```

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('monthly_car_sales_sample.csv',usecols=['Model','Brand','Fuel_Type','Transmission'])
df.head()
```

```python
%%markdown
---
Using Python
```

```python
df['Transmission'].value_counts()
```

```python
df['Transmission']= df['Transmission'].map({'Manual':1,'Automatic':2,'Semi-Auto':3})
```

```python
df
```

```python
df['Fuel_Type']=df['Fuel_Type'].map({'Petrol':1,'Diesel':2,'Hybrid':3,'Electric':4})
```

```python
df
```

```python
df['Transmission'].apply(lambda x: 1 if x == 'Manual' else 0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Fuel_Type',axis=1),df['Fuel_Type'],test_size=0.33,random_state=42)
```

```python
X_test
```

```python
%%markdown
---
Applying Label Encoder
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
le = LabelEncoder()
```

```python
y_train_scal = le.fit_transform(y_train)
y_test_scal = le.fit_transform(y_test)
```

```python
y_test_scal
```

```python
oe = OrdinalEncoder()
X_train_scale = oe.fit_transform(X_train)
X_test_scale = oe.fit_transform(X_test)
```
