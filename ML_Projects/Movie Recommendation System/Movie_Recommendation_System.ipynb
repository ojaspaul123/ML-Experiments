```python
df = pd.concat([df_netflix,df_amazon,df_HBO],axis =0,ignore_index=True)
```

```python
%%markdown
---
Data Cleaning and Preprocessing
```

```python
df_movies = df.drop_duplicates()
```

```python
df_movies.duplicated().sum()
```

```python
df_movies.drop(['description', 'age_certification'], axis=1, inplace=True)
```

```python
df_movies
```

```python
df['production_countries']
```

```python
df_movies['production_countries'] = df_movies['production_countries'].str.replace(r"\[", '', regex=True).str.replace(r"'", '', regex=True).str.replace(r"\]", '', regex=True)
```

```python
df_movies['lead_prod_country'] = df_movies['production_countries'].str.split(',').str[0]
```

```python
df_movies['prod_countries_cnt'] = df_movies['production_countries'].str.split(',').str.len()
```

```python
df_movies['lead_prod_country'] = df_movies['lead_prod_country'].replace('', np.nan)
```

```python
df_movies.shape
```

```python
df_movies['lead_prod_country']
```

```python
df_movies['genres']

```

```python
df_movies['genres'] = df_movies['genres'].str.replace(r"\[", '', regex=True).str.replace(r"'", '', regex=True).str.replace(r"\]", '', regex=True)
```

```python
df_movies['main_genre'] = df_movies['genres'].str.split(',').str[0]
```

```python
df_movies['main_genre'] = df_movies['main_genre'].replace('', np.nan)
```

```python
df_movies['main_genre']

```

```python
%%markdown
---
Drop Missing Values
```

```python
  df_movies.shape
```

```python
df_movies.isnull().sum()
```

```python
df_movies.drop(['id', 'imdb_id'], axis=1, inplace=True)
```

```python
df_movies.dropna(subset=['release_year', 'main_genre'], inplace=True)
print('Rows after targeted dropna:', len(df_movies))

# Impute remaining numeric NaNs with column median
num_cols = df_movies.select_dtypes(include='number').columns
df_movies[num_cols] = df_movies[num_cols].apply(lambda col: col.fillna(col.median()))
print('NaN remaining:', df_movies.isnull().sum().sum())
```

```python
df_movies.head()
```

```python
print(df_movies.columns.tolist())
print(df_movies.index.name)
```

```python
print(df_movies.shape)
```

```python
%%markdown
---
Encoding Categorical Features
```

```python
cat_cols = [c for c in ['type', 'main_genre'] if c in df_movies.columns]
dummies = pd.get_dummies(df_movies[cat_cols], drop_first=True)
df_num = df_movies.drop(columns=cat_cols)
df_movies_dum = pd.concat([df_num, dummies], axis=1)
print('Shape after encoding:', df_movies_dum.shape)
```

```python
df_movies_dum.shape
```

```python
df_movies_dum
```

```python
df_movies_dum = df_movies_dum.select_dtypes(include='number')
```

```python
%%markdown
---
Scaling (MinmaxScaler)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled_arr = scaler.fit_transform(df_movies_dum)
df_scaled = pd.DataFrame(df_scaled_arr, columns=df_movies_dum.columns, index=df_movies_dum.index)
df_scaled.head()
```
