```python
##keeping important columns(Feature_Scaling)
col_sort = ['gender','SeniorCitizen','Partner',	'Dependents','tenure','PhoneService'	,'MultipleLines','Contract','TotalCharges','Churn']
df = df[col_sort]
df.head()
```

```python
%%markdown
---
Encoding binary variables
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
col_trans = ['gender','Partner','Dependents','PhoneService','MultipleLines','Contract','Churn']
for col in col_trans:
  df[col] = label_encoder.fit_transform(df[col])
```
